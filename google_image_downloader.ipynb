 Step 2: Import and Define Download Function

```python
from icrawler.builtin import GoogleImageCrawler

def download_images(query, max_images=10, output_dir='downloaded_images'):
    # Set up crawler with custom folder
    google_crawler = GoogleImageCrawler(storage={'root_dir': f'{output_dir}/{query}'})

    # Start crawling
    google_crawler.crawl(keyword=query, max_num=max_images)

    print(f"✅ Downloaded {max_images} images for '{query}' in folder '{output_dir}/{query}'")
```

 Step 3: Run the Downloader


```python
# Input from user
query = input("Enter search query: ")
num_images = int(input("Enter number of images to download: "))

# Call the function
download_images(query, num_images)
```

Step 4: View Downloaded Images in Colab

```python
import os
from IPython.display import Image, display

def show_images_from_folder(folder_path, num_show=5):
    image_files = [f for f in os.listdir(folder_path) if f.endswith(('jpg', 'png', 'jpeg'))]

    for img_file in image_files[:num_show]:
        display(Image(filename=os.path.join(folder_path, img_file)))

# Example: show first 5 images from downloaded folder
show_images_from_folder(f'downloaded_images/{query}', num_show=5)
```
